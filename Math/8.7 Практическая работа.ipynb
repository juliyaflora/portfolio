```python
from sympy import *
from sympy.plotting import plot
init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x, y = symbols('x, y')
```

```python
f1 = x*(y-5)**2
f1
```

```python
f1_diff_x = diff(f1,x)
f1_diff_x
```

```python
f1_diff_y = diff(f1,y)
f1_diff_y
```

```python
nonlinsolve([f1_diff_x,f1_diff_y],[x,y])
```

```python
plot3d(f1_diff_y)
```

```python
plot3d(f1_diff_x)
```

### Задание 2

```python
a2, a1, a0 = symbols('a2, a1, a0')
```

```python
MSE = 1/3 * (((a2 * 1945 + a1 * 560 + a0) - 4.3)**2 + ((a2 * 1495 + a1 * 340 + a0) - 4.9)**2 + ((a2 * 1570 + a1 * 343 + a0) - 5.2)**2)
MSE
```

```python
MSEa2 = diff(MSE, a2)
MSEa2
```

```python
MSEa1 = diff(MSE, a1)
MSEa1
```

```python
MSEa0 = diff(MSE, a0)
MSEa0
```

```python
nonlinsolve([MSEa2, MSEa1, MSEa0], [a2, a1, a0])
```
